# State-machine interface
Pyplot provides the state-machine interface to the underlying object-oriented plotting library. The state-machine implicitly and automatically creates figures and axes to achieve the desired plot. For example:

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear', marker='o', markersize=4)
plt.plot(x, x**2, label='quadratic', marker='s', markersize=4)
plt.plot(x, x**3, label='cubic', marker='^', markersize=4)

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot with Markers")

plt.legend()

plt.show()
```

```python
# Importing necessary libraries
import numpy as np  # NumPy for numerical operations like creating arrays
import matplotlib.pyplot as plt  # Matplotlib for plotting graphs

# Generating 100 equally spaced numbers between 0 and 2 (for X-axis)
x = np.linspace(0, 2, 100)

# Plotting y = x (Linear function)
# 'marker' specifies the point marker style ('o' for circle)
# 'markersize' sets the size of the markers
plt.plot(x, x, label='linear', marker='o', markersize=2)

# Plotting y = x^2 (Quadratic function) with diamond markers
plt.plot(x, x**2, label='quadratic', marker='d', markersize=3)

# Plotting y = x^3 (Cubic function) with triangle-up markers
plt.plot(x, x**3, label='cubic', marker='^', markersize=4)

# Adding label to the X-axis
plt.xlabel('x label')  # You can change this to a more descriptive name if needed

# Adding label to the Y-axis
plt.ylabel('y label')  # You can also customize this label

# Adding a title to the plot
plt.title("Simple Plot with Markers")  # Describes what the plot shows

# Displaying the legend to identify each line (uses the 'label' from plt.plot)
plt.legend()

# Showing the final plot with all the above settings
plt.show()
```

# Working with NumPy arrays
Generally, we have to work with NumPy arrays. All sequences are converted to numpy arrays internally. The below example illustrates plotting several lines with different format styles in one command using arrays.

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
# Importing the required libraries
import numpy as np
import matplotlib.pyplot as plt

# Creating a NumPy array 't' representing time points.
# It generates values starting from 0 up to (but not including) 5,
# with a step of 0.2 (i.e., samples every 0.2 seconds).
t = np.arange(0., 5., 0.2)

# Plotting multiple datasets on the same graph using different styles:
# 1. Plot of y = t (linear) with red dashed lines ('r--')
# 2. Plot of y = t^2 (quadratic) with blue square markers ('bs')
# 3. Plot of y = t^3 (cubic) with green triangle-up markers ('g^')
plt.plot(
    t, t, 'r--',       # Red dashed line for y = t
    t, t**2, 'bs',     # Blue squares for y = t^2
    t, t**3, 'g^'      # Green triangles for y = t^3
)

# Displaying the plot
plt.show()
```

# Object-Oriented API

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)


# Call plot() method on the appropriate object
ax[0].plot(x1, np.sin(x1), 'b-')
ax[1].plot(x1, np.cos(x1), 'b-');
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data using NumPy
# 'x1' contains 100 values from 0 to 10 (for smooth curves)
x1 = np.linspace(0, 10, 100)

# First, create a figure with a grid of 2 subplots (stacked vertically)
# 'fig' is the entire figure; 'ax' is an array containing two Axes objects.
fig, ax = plt.subplots(2)  # 2 subplots arranged vertically (nrows=2 by default)

# Plot sine curve on the first subplot (ax[0])
ax[0].plot(x1, np.sin(x1), 'b-')  # 'b-' means blue solid line
ax[0].set_title('Sine Wave')      # Adding a title for clarity
ax[0].set_ylabel('Amplitude')     # Y-axis label

# Plot cosine curve on the second subplot (ax[1])
ax[1].plot(x1, np.cos(x1), 'b-')  # Blue solid line for cosine
ax[1].set_title('Cosine Wave')    # Adding a title for clarity
ax[1].set_xlabel('X-axis')        # X-axis label (only at bottom plot)
ax[1].set_ylabel('Amplitude')     # Y-axis label

# Display the complete plot with both subplots
plt.tight_layout()  # Automatically adjusts spacing to prevent overlap
plt.show()
```

# Objects and Reference
The main idea with the Object Oriented API is to have objects that one can apply functions and actions on. The real advantage of this approach becomes apparent when more than one figure is created or when a figure contains more than one subplot.

We create a reference to the figure instance in the fig variable. Then, we ceate a new axis instance axes using the add_axes method in the Figure class instance fig as follows:-

```python
fig = plt.figure()

x2 = np.linspace(0, 5, 10)
y2 = x2 ** 2

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x2, y2, 'g')

axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title('Graph');
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a new figure
fig = plt.figure()

# Generate data
x = np.linspace(0, 5, 10)  # 10 points between 0 and 5
y = x ** 2                 # Square of x

# Add axes to the figure: [left, bottom, width, height] in fraction of figure size
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Plot the data with green line ('g')
axes.plot(x, y, color='green')

# Set labels and title
axes.set_xlabel('x-axis')       # X-axis label
axes.set_ylabel('y-axis')       # Y-axis label
axes.set_title('Tittle-Graph')         # Title of the graph

# Display the plot
plt.show()
```

# Multiline Plots
Multiline Plots mean plotting more than one plot on the same figure. We can plot more than one plot on the same figure.
It can be achieved by plotting all the lines before calling show(). It can be done as follows:-

```python
x4 = range(1, 5)
plt.plot(x4, [xi*1.5 for xi in x4])
plt.plot(x4, [xi*3 for xi in x4])
plt.plot(x4, [xi/3.0 for xi in x4])
plt.show()
```

```python
# Import the required library
import matplotlib.pyplot as plt

# Create X-axis data (numbers from 1 to 4)
x4 = range(1, 5)

# Plotting multiple lines with labels, colors, and markers:
# Plot for y = 1.5 * x (Blue line with circle markers)
plt.plot(x4, [xi * 1.5 for xi in x4],color='blue', marker='o', linestyle='-',linewidth=2, label='y = 1.5 * x')

# Plot for y = 3 * x (Green line with square markers)
plt.plot(x4, [xi * 3 for xi in x4],color='green', marker='s', linestyle='--', linewidth=2,label='y = 3 * x')

# Plot for y = x / 3 (Red line with triangle markers)
plt.plot(x4, [xi / 3.0 for xi in x4],color='red', marker='^', linestyle='-.', linewidth=2,label='y = x / 3')

# Adding title and axis labels
plt.title("Multiple Line Comparison", fontsize=16)
plt.xlabel("X-axis (Input Values)", fontsize=12)
plt.ylabel("Y-axis (Computed Values)", fontsize=12)

# Adding a legend to identify lines
plt.legend(loc='upper left')

# Display the plot window
plt.show()
```

# Bar Graph
Bar charts display rectangular bars either in vertical or horizontal form. Their length is proportional to the values they represent. They are used to compare two or more values.

We can plot a bar chart using plt.bar() function

```python
data2 = [5. , 25. , 50. , 20.]
plt.bar(range(len(data2)), data2)
plt.show() 
```

```python
# Import required libraries
import matplotlib.pyplot as plt

# Define the data for bar heights
data2 = [5, 25, 50, 20]

# Define the positions of bars on the x-axis
x_positions = range(len(data2))  # Automatically generates [0, 1, 2, 3]

# Create a bar chart
plt.bar(
    x_positions,   # X-axis positions of the bars
    data2,         # Heights of the bars
    color='skyblue',  # Optional: bar color for better appearance
    edgecolor='black' ) # Optional: adds border to bars


# Add title and axis labels
plt.title("Bar Chart Example", fontsize=14)
plt.xlabel("Categories", fontsize=12)
plt.ylabel("Values", fontsize=12)

# Optional: Add category labels instead of numbers (if desired)
plt.xticks(ticks=x_positions, labels=["A", "B", "C", "D"])

# Show the plot
plt.show()
```

#  Error Bar Chart

```python
x9 = np.arange(0, 4, 0.2)
y9 = np.exp(-x9)
e1 = 0.1 * np.abs(np.random.randn(len(y9)))
plt.errorbar(x9, y9, yerr = e1, fmt = '.-')
plt.show();
```

```python
#  Import required libraries
import numpy as np
import matplotlib.pyplot as plt

#  Generate X-axis values (from 0 to 4 with step of 0.2)
x9 = np.arange(0, 4, 0.2)

#  Compute Y-axis values using exponential decay function
y9 = np.exp(-x9)

#  Generate random error values (positive, scaled by 0.1)
e1 = 0.1 * np.abs(np.random.randn(len(y9)))

# Create an error bar plot
plt.errorbar(
    x9,           # X-axis data
    y9,           # Y-axis data
    yerr=e1,      # Y-axis error values
    fmt='.-',     # Marker and line style: dot markers with solid lines
    color='blue',  # Optional: Color for the line and markers
    ecolor='red',  # Optional: Color for error bars
    capsize=3 )     # Optional: Adds caps to error b

# Add title and axis labels for clarity
plt.title("Error Bar Plot Example", fontsize=14)
plt.xlabel("X-axis (Values from 0 to 4)", fontsize=12)
plt.ylabel("Y-axis (exp(-x))", fontsize=12)

# Display the plot
plt.show()
```

We can draw stacked bar chart by using a special parameter called bottom from the plt.bar() function.

```python
A = [15., 30., 45., 22.]
B = [15., 25., 50., 20.]
z2 = range(4)
plt.bar(z2, A, color = 'b')
plt.bar(z2, B, color = 'r', bottom = A)
plt.show()
```

```python
# Import required library
import matplotlib.pyplot as plt

# Define the heights for the first set of bars
A = [15, 30, 45, 22]

# Define the heights for the second set of bars (stacked on top of A)
B = [15, 25, 50, 20]

# Define positions for the bars on the X-axis
z2 = range(4)  # [0, 1, 2, 3]

# Create the stacked bar chart
plt.bar(
    z2,          # X-axis positions
    A,           # Heights of the first set of bars
    color='blue',  # Color for first set of bars
    label='Group A')  # Legend label for Group A


plt.bar(
    z2,           # X-axis positions (same as above for stacking)
    B,            # Heights of the second set of bars
    color='red',  # Color for second set of bars
    bottom=A,     # Stack B bars on top of A bars
    label='Group B')  # Legend label for Group B


# Add title and axis labels
plt.title("Stacked Bar Chart Example", fontsize=14)
plt.xlabel("Categories", fontsize=12)
plt.ylabel("Values", fontsize=12)

# Add legend for clarity
plt.legend()

# Show the plot
plt.show()
```
